```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('Downloads/diabetes_model.csv')
data.head()
```

# 1. Create a Count plot of Outcome

```python
sns.countplot(data['Outcome'])
```

# 2. Analyze the effect of age on Outcome using Boxplot

```python
sns.boxplot(x='Outcome',y='Age',data=data)
```

# 3. Check for the chances of getting diabetes if it is common in family, using the Daiabetes Pedigree Function

```python
sns.boxplot(x='Outcome',y='DiabetesPedigreeFunction',data=data)
plt.title('Diabetes Pedigree vs Outcome')
```

# 4. Check for the relationship with Insulin & Glucose and group the datapoints by Outcome.

```python
data.groupby(["Outcome"])[["Insulin","Glucose"]].count()
```

```python
sns.scatterplot(data['Insulin'],data['Glucose'])
plt.xlabel('Insulin')
plt.ylabel('Glucose')
```

# 5. Create a heatmap on Correlation of features in the data.

```python
data_heatmap=data[['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age','Outcome']]
sns.heatmap(data_heatmap.corr(),annot=True)
plt.xticks(rotation=45)
```
